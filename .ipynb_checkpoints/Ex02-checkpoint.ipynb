```python
import pandas as pd
import numpy as np
import math
```

```python
## Read data from source
```

```python
file = 'data.xlsx'
df = pd.read_excel(file, header=None)
nr_rows = df.shape[0]
nr_cols = df.shape[1]
df.head()

```

```python
## Convert feature to number
def numberize(value): 
    return 1 if value == 'M' else 0

df[0] = df[0].apply(lambda x: numberize(x))

    
# ## Normalize data
def norm(vector):
    min = np.amin(vector)
    max = np.amax(vector)
    return (vector - min) / (min - max )

for i in range(nr_cols):
    df[:][i] = norm(df[:][i])
    
df.head()    
```

```python
## Add '1' - Feature
```

```python
ones = np.ones(nr_rows, dtype=np.int)
df.insert(0, '', ones)
df.columns = range(df.shape[1])
df
```

```python
def sig(z):
    return 1 / (1 + math.e ** (-1*z))

def h(theta, x):
	z = 0
	for i in xrange(len(theta)):
		z += x[i]*theta[i]
	return sig(z)
```
